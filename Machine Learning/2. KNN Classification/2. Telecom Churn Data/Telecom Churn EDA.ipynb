### Convert Churned value to numeric data type

```python
churn_map = {
    True: 1,
    False: 0
}
data.churned = data.churned.map(churn_map)
data.churned.head()
```

```python
data.churned.unique()
```

### Describing a correlation between different columns

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(20,15))
sns.heatmap(data=data.corr(), annot=True, linecolor='w', linewidths=1)
plt.show()
```

### Which state has the highest number of churned users?

```python
state_users_churned = data.groupby(['state'])['churned'].value_counts()
state_users_churned
```

```python
state_users_churned.sort_values(ascending=False)
```
