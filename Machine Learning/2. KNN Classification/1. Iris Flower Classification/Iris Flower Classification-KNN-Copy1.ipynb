# Iris Flower Classification

```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
```

```python
data = load_iris()
```

```python
data
```

```python
training_data = data.data
```

```python
training_data
```

```python
columns = data.feature_names
```

```python
columns
```

```python
training_data = pd.DataFrame(training_data)
```

```python
training_data.columns = columns
```

```python
training_data.head()
```

```python
data.target_names
```

```python
training_data['target'] = data.target
```

```python
training_data.head()
```

### Performing Machine Learning - KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
output_data = training_data['target']
```

```python
training_data = training_data[columns[:4]]
training_data.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X, x_test, Y, y_test = train_test_split(training_data, 
                                        output_data, 
                                        test_size=0.2, 
                                        random_state = 42)
```

```python
model = KNeighborsClassifier(n_neighbors=10)
```

```python
model.fit(X, Y)
```

```python
prediction = model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion = confusion_matrix(prediction, y_test)
confusion
```

```python
score = accuracy_score(prediction, y_test)
score
```

```python
new_prediction = model.predict([[4.9,2.0,1.4,0.2]])
new_prediction
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_model = GridSearchCV(KNeighborsClassifier(), 
                          param_grid={'n_neighbors' : [1, 8,10, 20, 30, 50]})
```

```python
grid_model.fit(X, Y)
```

```python
grid_model.cv_results_
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```
